```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_v2.csv')
df.head()
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df['rating'].value_counts(normalize=True)
```

```python
df_user_based = df.pivot('user', 'movie', 'rating')
df_user_based.head()
```

### To make the example easier, let's just focus on rows that have 500 columns without Nan

```python
df_user_based = df_user_based.dropna(thresh=500)
df_user_based
```

```python
def mean_normalization(row):
  mean = row.mean()

  return row - mean

df_user_based_normalized = df_user_based.apply(mean_normalization, axis=1)
df_user_based_normalized.head()
```

```python
df_user_based_normalized.fillna(0, inplace=True)
```

```python
n_users = df_user_based_normalized.shape[0]

triangular_similarity_matrix = pd.DataFrame(np.nan, index=df_user_based_normalized.index, columns=df_user_based_normalized.index)
triangular_similarity_matrix.head()
```

```python
def cosine_similarity(vector_a, vector_b):
  scalar_product = 0
  magnitude_a = 0
  magnitude_b  = 0

  for i in range(len(vector_a)):
    scalar_product += vector_a[i] * vector_b[i]

    magnitude_a += vector_a[i] ** 2
    magnitude_b += vector_b[i] ** 2

  magnitude_a = magnitude_a ** (1/2)
  magnitude_b = magnitude_b ** (1/2)

  if magnitude_a == 0 or magnitude_b == 0:
    return 0

  return scalar_product / (magnitude_a * magnitude_b)

for current_user in range(n_users):
  vector_a = df_user_based_normalized.iloc[current_user, :].values

  for aux_user in range(current_user, n_users):
    vector_b = df_user_based_normalized.iloc[aux_user, :].values

    triangular_similarity_matrix.iloc[current_user, aux_user] = cosine_similarity(vector_a, vector_b)
```

### Let's make the matrix symmetric to simplify the example.

```python
for i in range(triangular_similarity_matrix.shape[0]):
  for j in range(i):
    triangular_similarity_matrix.iloc[i, j] = triangular_similarity_matrix.iloc[j, i]
```

```python
user_id = np.random.choice(df_user_based_normalized.index)

movies_not_watched_by_user = df_user_based_normalized.loc[user_id][df_user_based_normalized.loc[user_id] == 0].index

movie_id = np.random.choice(movies_not_watched_by_user)

user_id, movie_id
```

```python
users_watched_movie = df_user_based_normalized[df_user_based_normalized.loc[:, movie_id] != 0].index
users_watched_movie
```

```python
users_similarity = triangular_similarity_matrix.loc[user_id, users_watched_movie]

total_weight = sum(users_similarity)
total_sum = 0

for i, value in users_similarity.items():
  total_sum += value * df_user_based_normalized.loc[i, movie_id]

weighted_mean = total_sum / total_weight
recommendation = movie_id if weighted_mean >= 0.5 else f'{movie_id} not'

print(f'Weighted Mean: {weighted_mean:.3f}')
print(f'Movie {recommendation} recommended for user {user_id}')
```
